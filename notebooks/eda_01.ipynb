```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
irr_data = pd.read_csv ("../dataset/train.csv")
```

```python
irr_data
```

```python
irr_data.describe()
```

```python
class_mapping = {"Low": 0, "Medium": 1, "High": 2}
irr_data ["class_enc"] = irr_data["Irrigation_Need"].map(class_mapping)
```

```python
corr = irr_data.select_dtypes(include="number").corr()
sns.heatmap(corr, cmap="YlGnBu")
plt.show()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
splitobj = StratifiedShuffleSplit (n_splits=1, test_size= 0.2)
for trainidx, testidx in splitobj.split(irr_data, irr_data["class_enc"]):
    strat_train_set = irr_data.loc[trainidx]
    strat_val_set = irr_data.loc[testidx]
```

```python
strat_train_set.to_pickle("strat_train_set")
strat_val_set.to_pickle("strat_val_set")
```
